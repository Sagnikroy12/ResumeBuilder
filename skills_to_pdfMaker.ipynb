```python
import json

def get_input(prompt, optional=False):
    val = input(prompt)
    return val if val or not optional else None

def main():
    data = {}

    # 1–7: Basic info
    data["name"] = get_input("1. Name: ")
    data["address"] = get_input("2. Address: ")
    data["phone"] = get_input("3. Phone number: ")
    data["email"] = get_input("4. Email: ")
    data["linkedin"] = get_input("5. LinkedIn: ")
    data["portfolio"] = get_input("6. Portfolio (optional): ", optional=True)
    data["github"] = get_input("7. GitHub (optional): ", optional=True)

    # 8: Objective
    data["objective"] = get_input("8. Objective: ")

    # 9: Skills
    data["skills"] = {
        "programming_languages": get_input("9.a Programming languages (comma-separated): ").split(","),
        "tools": get_input("9.b Tools (comma-separated): ").split(","),
        "skills": get_input("9.c Other skills (comma-separated): ").split(",")
    }

    # 10: Experience
    data["experience"] = []

    latest_exp = {
        "role": get_input("10.a.I Role: "),
        "company": get_input("10.a.II Company: "),
        "duration": get_input("10.a.III Duration: "),
        "achievements": get_input("10.a.IV Promotions and achievements (comma-separated): ").split(",")
    }
    data["experience"].append(latest_exp)

    if get_input("Do you want to add second experience? (y/n): ").lower() == "y":
        second_exp = {
            "role": get_input("10.b.I Role: "),
            "company": get_input("10.b.II Company: "),
            "duration": get_input("10.b.III Duration: "),
        }
        data["experience"].append(second_exp)

    # 11: Projects
    projects = []
    while get_input("Do you want to add a project? (y/n): ").lower() == "y":
        projects.append({
            "name": get_input("Project name: "),
            "description": get_input("Project description: ")
        })
    data["projects"] = projects

    # 12: Education
    data["education"] = get_input("12. Education: ")

    # 13: Certification
    data["certifications"] = get_input("13. Certifications (comma-separated): ").split(",")

    # Save to JSON
    with open("resume_data.json", "w") as f:
        json.dump(data, f, indent=4)

    print("\n✅ Data collected and saved to resume_data.json!")

if __name__ == "__main__":
    main()
```

```python
import json
from fpdf import FPDF

class PDF(FPDF):
    def header(self):
        self.set_font("Arial", "B", 14)
        self.cell(0, 10, "Resume", ln=True, align="C")

    def add_section(self, title):
        self.set_font("Arial", "B", 12)
        self.ln(5)
        self.cell(0, 10, title, ln=True)

    def add_multiline(self, label, value):
        self.set_font("Arial", "B", 11)
        self.cell(30, 10, f"{label}:", ln=False)
        self.set_font("Arial", "", 11)
        self.multi_cell(0, 10, value)

def generate_pdf(json_path, pdf_path):
    with open(json_path, "r") as f:
        data = json.load(f)

    pdf = PDF()
    pdf.add_page()

    # Name and contact
    pdf.set_font("Arial", "B", 16)
    pdf.cell(0, 10, data["name"], ln=True)
    pdf.set_font("Arial", "", 11)
    pdf.cell(0, 10, f"{data['email']} | {data['phone']} | {data['linkedin']}", ln=True)
    if data.get("portfolio"):
        pdf.cell(0, 10, f"Portfolio: {data['portfolio']}", ln=True)
    if data.get("github"):
        pdf.cell(0, 10, f"GitHub: {data['github']}", ln=True)

    # Objective
    pdf.add_section("Objective")
    pdf.multi_cell(0, 10, data["objective"])

    # Skills
    pdf.add_section("Skills")
    skills = data["skills"]
    pdf.multi_cell(0, 10, f"Programming Languages: {', '.join(skills['programming_languages'])}")
    pdf.multi_cell(0, 10, f"Tools: {', '.join(skills['tools'])}")
    pdf.multi_cell(0, 10, f"Other Skills: {', '.join(skills['skills'])}")

    # Experience
    pdf.add_section("Experience")
    for exp in data["experience"]:
        pdf.set_font("Arial", "B", 11)
        pdf.cell(0, 10, f"{exp['role']} at {exp['company']} ({exp['duration']})", ln=True)
        if exp.get("achievements"):
            pdf.set_font("Arial", "", 11)
            pdf.multi_cell(0, 10, f"Achievements: {', '.join(exp['achievements'])}")

    # Projects
    if data.get("projects"):
        pdf.add_section("Projects")
        for proj in data["projects"]:
            pdf.set_font("Arial", "B", 11)
            pdf.cell(0, 10, proj["name"], ln=True)
            pdf.set_font("Arial", "", 11)
            pdf.multi_cell(0, 10, proj["description"])

    # Education
    pdf.add_section("Education")
    pdf.multi_cell(0, 10, data["education"])

    # Certification
    if data.get("certifications"):
        pdf.add_section("Certifications")
        pdf.multi_cell(0, 10, ", ".join(data["certifications"]))

    pdf.output(pdf_path)
    print(f"✅ Resume PDF generated: {pdf_path}")

# Run this to convert JSON to PDF
generate_pdf("resume_data.json", "Sagnik_Resume.pdf")
```
